# Analyzing data on Thanksgiving dinner in the US

```python
import pandas as pd
data=pd.read_csv("thanksgiving-2015-poll-data.csv",encoding="Latin-1")
data.head(2)
```

```python
data.columns.values
```

```python
q1="Do you celebrate Thanksgiving?"
count_celebrate=pd.value_counts(data[q1])
count_celebrate
```

```python
celebrate_yes=data[data[q1]=="Yes"]
celebrate_yes.head(10)
```

```python
q2="What is typically the main dish at your Thanksgiving dinner?"
count_dish=data[q2].value_counts(ascending=False)
count_dish
```

```python
dish_Tofurkey=data[data[q2]=="Tofurkey"]
q3="Do you typically have gravy?"
Tofurkey_w_gravy=dish_Tofurkey[q3]
Tofurkey_w_gravy.value_counts()
```

Figuring Out What Pies People Eat
Where ate_pies is True, they ate none of the types of pies.

```python
q4="Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Apple"
q5="Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pumpkin"
q6="Which type of pie is typically served at your Thanksgiving dinner? Please select all that apply. - Pecan"
apple_isnull=pd.isnull(data[q4])
pumpkin_isnull=pd.isnull(data[q5])
pecan_isnull=pd.isnull(data[q6])
ate_pies=(apple_isnull&pumpkin_isnull&pecan_isnull)
pd.value_counts(ate_pies)
```

Converting Age To Numeric
The result is assigned to the int_age column of data.

```python
def age_convert(string):
    if pd.isnull(string):
        return None
    else:
        age=string.split(" ")[0]
        age=age.replace("+","")
        return int(age)

int_age=data["Age"].apply(age_convert)
data["int_age"]=int_age
data["int_age"].describe()
```

Converting Income to numeric
The result is assigned to the int_income column of data.

```python
def str_2_int(string):
    if pd.isnull(string):
        return None
    else:
        first_item=string.split(" ")[0]
        if first_item=="Prefer":
            return None
        else:
            first_item=first_item.replace(",","")
            first_item=first_item.replace("$","")
            return int(first_item)
        
q7="How much total combined money did all members of your HOUSEHOLD earn last year?"

int_income=data[q7].apply(str_2_int)
data["int_income"]=int_income
data["int_income"].describe()
```

Correlating Travel Distance and Income

```python
q8="How far will you travel for Thanksgiving?"
travel_less_150000=data[data["int_income"]<150000][q8]
print(pd.value_counts(travel_less_150000))
travel_over_150000=data[data["int_income"]>=150000][q8]
print(pd.value_counts(travel_over_150000))
```
